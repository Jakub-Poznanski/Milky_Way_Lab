```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm
import matplotlib.colors as cm
from astropy.io import fits
from astropy.wcs import WCS
from astropy import units as u
from matplotlib.gridspec import GridSpec
import astropy.io.fits as pyfits
from astropy.nddata import Cutout2D
from astropy.coordinates import SkyCoord
from astropy.wcs.utils import skycoord_to_pixel
import matplotlib.colors as mc
from astropy.table import Table, vstack
```

```python
def open_pv(file_path, coordinate_type = 'longitude'):
    """
    Opens a FITS file containing a position-velocity plot and displays it.
    
    Args:
    File_Path (str): Path to the FITS file
    coordinate_type (str): Type of position coordinate, either 'longitude' or 'latitude'
    
    Returns:
    None (displays the plot)
    """
    
    with fits.open(file_path) as hdul:
        image_data = hdul[0].data
        header = hdul[0].header

    wcs = WCS(header)

    fig, ax = plt.subplots(figsize = (10, 8), subplot_kw = {'projection': wcs})

    im = ax.imshow(image_data, cmap = 'magma', origin = 'lower', aspect = 'auto')

    if coordinate_type.lower() == 'longitude':
        ax.set_xlabel('Galactic Longitude')
        lon = ax.coords['glon']
        lon.set_major_formatter('d.dd')
    elif coordinate_type.lower() == 'latitude':
        ax.set_xlabel('Galactic Latitude')
        lat = ax.coords['glat']
        lat.set_major_formatter('d.dd')
        
    else:
        raise ValueError("coordinate_type must be either 'longitude' or 'latitude'")

    ax.set_ylabel('Velocity m_s^-1')
    
    plt.colorbar(im, ax = ax, label = 'Intensity')

    plt.title('Position-Velocity Plot')

    plt.show()
```
